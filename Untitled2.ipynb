#Aplikasi Sederhana menghitung Volume Bola

```python
r = float(input("Masukkan jari-jari bola (dalam satuan cm): "))
pi = 3.14
volume = (4/3) * pi * r**3

print(f"Volume bola dengan jari-jari {r} cm adalah {volume:.2f} cm³")
```
