#Latihan Soal: Harga Barang di Swalayan

##1. Single Assigment

```python
susu = 15000
roti = 12000
telur = 24000


print(f" susu = Rp{susu:,}")
print(f" roti = Rp{roti:,}")
print(f" telur = Rp{telur:,}\n")
```

##2. Multiple Assigment

```python
gula, minyak = 14000, 20000
print(f" gula = Rp{gula:,}")
print(f" minyak = Rp{minyak:,}\n")
```

##3. Pengisian Nilai Sama

```python
mie_ayam = mie_soto = mie_kari = 3500
print(f" mie_ayam = Rp{mie_ayam:,}")
print(f" mie_soto = Rp{mie_soto:,}")
print(f" mie_kari = Rp{mie_kari:,}")
```

##4. Swap (Tukar Nilai)

```python
susu, roti = roti, susu
print(f" susu = Rp{susu:,},\n roti = Rp{roti:,}")
```

##5. Perubahan Nilai Variabel

```python
print(f"  telur (sebelum) = Rp{telur:,}")
telur -= 2000  # harga telur turun Rp2.000
print(f"  telur (sesudah) = Rp{telur:,} (turun Rp2.000)")
print(f"  minyak (sebelum) = Rp{minyak - 3000:,}")
minyak += 3000  # harga minyak naik Rp3.000
print(f"  minyak (sesudah) = Rp{minyak:,} (naik Rp3.000)\n")
```
